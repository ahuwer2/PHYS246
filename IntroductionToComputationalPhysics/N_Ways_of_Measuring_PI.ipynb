In this exercise we will compute the value of $\pi$ using a series (a sum of bunch of terms). We will use different series which converge at different rates.

### a. $\pi$ from $\tan$

Recall that we can generally find infinite series representations of transcendental functions like $\sin(x)$. In particular

$$\tan^{-1}(x) = \sum_{n=0}^\infty \frac{(-1)^n}{2n+1} x^{2n+1} ; -1<x\leq 1$$

Since $\tan^{-1}(1) = \pi/4$, we can write the following (slowly converging) infinite series:

$$\frac{\pi}{4} = \sum_{n=0}^\infty \frac{(-1)^n}{2n+1} = 1 - \frac{1}{3} + \frac{1}{5} - \frac{1}{7} + ...$$

If we group adjacent terms in the series we can rewrite this as

$$
\begin{align}
\frac{\pi}{4} = \sum_{n=0}^\infty \frac{(-1)^n}{2n+1} &= \left(1-\frac{1}{3}\right) + \left(\frac{1}{5}-\frac{1}{7}\right) + \left(\frac{1}{9}-\frac{1}{11}\right) + ...\\
&=\frac{3-1}{3 \cdot 1} + \frac{7-5}{7 \cdot 5} + \frac{11-9}{11 \cdot 9}+...\\
&=\frac{2}{3}+ \frac{2}{35}+\frac{2}{99} + ...\\
&= \boxed{2 \cdot \sum_{n=0}^\infty \left[ \frac{1}{(4n+3)(4n+1)} \right]}
\end{align}
$$

The value of $\pi$ is $3.14159265358979323846264338327950288419716939937510582...$ though the precision with which your computer can calculate it is probably limited to fewer digits than this.

### c. $\pi$ from Ramanujan

A much more rapidly converging series was discovered by the brilliant Indian mathematician Srinivasa Ramanujan.  It is

$$\frac{1}{\pi} = \frac{2\sqrt{2}}{9801} \sum_{k=0}^\infty \frac{(4k)!(1103+26390k)}{(k!)^4396^{4k}}$$

where $k!$ ("k factorial") is $1\times 2\times 3\times ... \times k$ and $(4k)!$ is $1\times2\times3\times ... \times 4k$

🦉Write a Python script that calculates an approximation to $\pi$ using the Ramanujan series, and comment on its accuracy after 1, 2, and 3 terms. (Recall that the value of $\pi$ is 3.14159265358979323846264338327950288419716939937510582...)

Note that there are even faster-converging formulas than this! One, mentioned in Wolfram MathWorld[<sup>1</sup>](#fn1), adds 50 digits of precision for each additional term.

### c. Python Fun - List Comprehensions

A cool Python feature is list comprehensions. Instead of writing a `for` loop, if we're clever we can stuff everything into one line inside a list, and python will be able to fill out the list faster than doing an `append()`.

### d.  Inscribed Polygon (EC) (Extra Credit: 5 points)

Do the same thing you did earlier but use both the inscribed and circumscribed polygons.  On your plot, you should see that you approach $\pi$ from above and below.

<div><img src="https://clark.physics.illinois.edu/246img/AnsStart.svg" width=200 align=left alt="Answer (start)"></img><br></div>

````python
# ANSWER HERE
def DrawNPolyInsc(n):
  dTheta=2*np.pi/n
  r=1
  thetas=[i*dTheta for i in range(n+1)]
  xs=[r*math.cos(theta) for theta in thetas]
  ys=[r*math.sin(theta) for theta in thetas]
  dy=(ys[1]-ys[0])
  dx=(xs[1]-xs[0])
  perim=math.sqrt(dy**2+dx**2)*n
  # Convert to np array
  xs=np.array(xs)
  ys=np.array(ys)
  # plot
  plt.plot(xs,ys)
  return perim/2

def DrawNPolyCirc(n):
  dTheta=2*np.pi/n
  # radius is no longer one, by geometry we must change it to the below value
  # which is dependent on the number of sides n.
  r=1/math.cos(dTheta/2)
  thetas=[i*dTheta for i in range(n+1)]
  xs=[r*math.cos(theta) for theta in thetas]
  ys=[r*math.sin(theta) for theta in thetas]
  dy=(ys[1]-ys[0])
  dx=(xs[1]-xs[0])
  perim=math.sqrt(dy**2+dx**2)*n
  # Convert to np array
  xs=np.array(xs)
  ys=np.array(ys)
  # plot
  plt.plot(xs,ys)
  return perim/2

halfPerimsInsc=[]
halfPerimsCirc=[]
DrawCircle()
for n in range(3,101):
  halfPerimInsc=DrawNPolyInsc(n)
  halfPerimCirc=DrawNPolyCirc(n)
  halfPerimsInsc.append(halfPerimInsc)
  halfPerimsCirc.append(halfPerimCirc)
plt.axis('scaled') # this makes your circle look like a circle and not an oval
plt.show()
plt.plot([n for n in range(3,101)],halfPerimsInsc)
plt.plot([n for n in range(3,101)],halfPerimsCirc)
plt.axhline(np.pi,linestyle='--')
plt.xlabel("Number of Sides")
plt.ylabel("Estimated Pi Value")
plt.title("Pi Estimation")
plt.legend(['Inscribed','Circumscribed'])
plt.show()
````

### b. $\pi$ from Darts

🦉Within your loops, measure the fraction of dots that actually end up in your circle.  Measure $\pi$ using this number both with 25 dots as well as 2500 dots (for this latter number you might want to turn off the plotting or it will be really slow).

*Code for fraction of points which are in the circle.  This should be $\pi/4$ so multiply by 4 to get $\pi$*
<div><img src="https://clark.physics.illinois.edu/246img/AnsStart.svg" width=200 align=left alt="Answer (start)"></img><br></div>

````python
# ANSWER HERE
DrawCircle()
def dartboard(n,plot=False):
  numWithin=0
  for i in range(0,n):
    x=random.uniform(-1,1)
    y=random.uniform(-1,1)
    if (x**2+y**2<1):
      numWithin+=1
    if plot:
      plt.plot(x,y,'.',color='red')
  return numWithin/n*4

piEst25=dartboard(25,True)
piEst2500=dartboard(2500)

plt.xlim(-1,1)
plt.ylim(-1,1)
plt.gca().get_xaxis().set_visible(False)
plt.gca().get_yaxis().set_visible(False)
plt.axis('scaled')
plt.show()

print(piEst25)
print(piEst2500)
````

````python
from matplotlib.patches import Rectangle
# Use t for the width of strips along x-axis,
# so width var should be divisible by t
_,ax=plt.subplots()
t=5
width=500
height=500
plt.xlim((0,width))
plt.ylim((0,height))
for i in range(0,width//t,2):
  ax.add_patch(Rectangle((t*i, 0), t, height))
# List of x positions of borders
linepos=[t*(i) for i in range(0,width//t) if not i==0]

def dropNeedle(l, plot=False):
  angle=random.uniform(0,2*np.pi)
  # Use l/2 here so that we measure relative to the center of the stick
  relx=l/2*math.cos(angle)
  rely=l/2*math.sin(angle)
  valid=False
  while not valid:
    xpos=random.uniform(0,width)
    ypos=random.uniform(0,height)
    # These are the positions of the ends, used to make sure it's in bounds
    posCheck=([xpos+relx,xpos-relx],[ypos+rely,ypos-rely])
    # Long if statement, just checking in bounds or not and retrying if not
    if all(x > 0 for x in posCheck[0]) and all(x < width for x in posCheck[0]) and all(y > 0 for y in posCheck[1]) and all(y < height for y in posCheck[1]):
      valid=True
  if plot:
    plt.plot(posCheck[0],posCheck[1],color='black')
  # Checking if it crosses a line
  for pos in linepos:
    if posCheck[0][0] > posCheck[0][1]:
      if posCheck[0][1] < pos and pos < posCheck[0][0]:
        return True
    else:
      if posCheck[0][0] < pos and pos < posCheck[0][1]:
        return True
  return False

timesCrossed=0
timesToCheck=20000
# Choose l<t
l=3
for i in range(timesToCheck):
  # set 2nd arg to True to display sticks, though beware the performance implications
  # For example, 20000 timesToCheck takes around 20s
  if dropNeedle(l, False) == True:
    timesCrossed+=1

# Using the equation p=2*l/(t*pi) for the l<t case
prob=timesCrossed/timesToCheck
piEstimate=(2*l)/(prob*t)
print('Estimate for pi:',piEstimate)
plt.show()




````
